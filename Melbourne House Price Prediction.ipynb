```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import linear_model
from sklearn.linear_model import Ridge
```

```python
df=pd.read_csv("/content/drive/MyDrive/Melbourne_housing_FULL.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
# Count Unique Values
df.nunique()
```

```python
#  Use Limited Columns makes more Sense
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
               'Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=df[cols_to_use]
```

```python
df.shape
```

```python
# Checking for NaN Values
df.isna().sum()
```

```python
# Handling Missing Values
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom', 'Car']
df[cols_to_fill_zero]=df[cols_to_fill_zero].fillna(0)
df ['Landsize'] = df ['Landsize']. fillna(df. Landsize.mean())
df ['BuildingArea'] = df ['BuildingArea']. fillna(df.BuildingArea.mean())
```

```python
df
```

```python
# Checking for NaN Values
df.isna().sum()
```

```python
# Drop NaN Values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df
```

```python
# One Hot Encode Categorical Features
df=pd.get_dummies(df,drop_first=True).astype(int)
```

```python
df.shape
```

```python
df
```

```python
df.head()
```

```python
# Split Data Into Training and Testing
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
reg=LinearRegression().fit(X_train,y_train)
```

```python
y_pred_train=reg.predict(X_train)
y_pred_test=reg.predict(X_test)
```

```python
# Score of Testing Data
r2_score(y_test,y_pred_test)
```

```python
# Score of Training Data
r2_score(y_train,y_pred_train)
```

```python
lasso_reg=linear_model.Lasso(alpha=50)
lasso_reg.fit(X_train,y_train)
```

```python
# Score of Testing Data
lasso_reg.score(X_test,y_test)
```

```python
# Score of Training Data
lasso_reg.score(X_train,y_train)
```

```python
ridge_reg=Ridge(alpha=50)
ridge_reg.fit(X_train,y_train)
```

```python
# Score of Testing Data
ridge_reg.score(X_test,y_test)
```

```python
# Score of Training Data
ridge_reg.score(X_train,y_train)
```
